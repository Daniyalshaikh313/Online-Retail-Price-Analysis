```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
flipkart = pd.read_csv("Flipkart laptops_info.csv", encoding='ISO-8859-1', on_bad_lines='skip')
amazon = pd.read_csv("Amazon Laptop Data.csv", encoding='ISO-8859-1', on_bad_lines='skip')
```

```python
flipkart.head()
```

```python
# Droping the column Name :
df1 = flipkart.drop(columns=['Platform', 'Reviews'])
df1
```

```python
# Changing the Columns Name to Above match name 
df = amazon.rename(columns={
    'Product Name': 'ProductName',
    'Original Price (in Rs.)': 'MRP',
    'Final Price (in Rs.)': 'Current_Price',
    'Ratings': 'Stars',
    'Total Ratings count': 'Ratings'
})
```

```python
# Droping the Columns Name :
df2 = df.drop(columns=['Discount Percentage'], errors='ignore')
df2
```

```python
# Imputing the 0 value into Rating Columns with Respect to NO Ratings : 
df1['Rating'] = df1['Ratings'].replace('No Ratings', 0)

# Convert the 'Rating' column to numeric if you need it as a number
df1['Ratings'] = pd.to_numeric(df1['Ratings'], errors='coerce')
df1
```

```python
## Imputing the 0 value into Rating Columns with Respect to NO Ratings : 
df2['Ratings'] = df1['Ratings'].replace('No Ratings', 0)

# Convert the 'Rating' column to numeric if you need it as a number
df2['Ratings'] = pd.to_numeric(df1['Ratings'], errors='coerce')
df2

df2 = df2.drop(columns=['Rating','Discount Percentage'], errors='ignore')
df2
```

```python
## Imputing the 0 value into Rating Columns with Respect to NO Ratings : 
df2['Stars'] = df1['Stars'].replace('No Ratings', 0)

# Convert the 'Rating' column to numeric if you need it as a number
df2['Stars'] = pd.to_numeric(df1['Stars'], errors='coerce')
df2
```

```python
# Replace 'NO rating' with NaN in Ratings column and convert to numeric
df2['Ratings'] = df2['Ratings'].replace('NO rating', None)
df2['Ratings'] = pd.to_numeric(df2['Ratings'], errors='coerce')

# Fill NaN values in 'Stars' and 'Ratings' columns with 0
df2['Stars'] = df2['Stars'].fillna(0)
df2['Ratings'] = df2['Ratings'].fillna(0)
df2
```

```python
# 1. Distribution of Ratings and Stars
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df1['Stars'], kde=True, color='blue', bins=5)
plt.title('Distribution of Stars')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.show()

# Amazon 
plt.figure(figsize=(14, 6))
sns.histplot(df2['Stars'], kde=True, color='blue', bins=5)
plt.title('Distribution of Stars')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate correlation matrix for numeric columns
correlation_matrix = df1[['Stars', 'Ratings', 'Current_Price', 'MRP']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Columns')
plt.show()
```
